#grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) #diagonal 
g.map_upper(plt.scatter) #upper side
g.map_lower(sns.kdeplot) #lower side
```

```python
n = sns.FacetGrid(data=tips,col='time',row='smoker')
n.map(sns.distplot,'tip')
```

```python
n = sns.FacetGrid(data=tips,col='time',row='smoker')
n.map(sns.scatterplot,'total_bill','tip')
```
